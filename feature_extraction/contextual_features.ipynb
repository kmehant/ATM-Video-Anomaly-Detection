```python
import cv2
(major_ver, minor_ver, subminor_ver) = (cv2.__version__).split('.')


def get_frames(video_location, dest, rfps):
    cap = cv2.VideoCapture(video_location)
    fps = 0
    if int(major_ver)  < 3 :
        fps = cap.get(cv2.cv.CV_CAP_PROP_FPS)
    else :
        fps = cap.get(cv2.CAP_PROP_FPS)
    i = 0
    # a variable to set how many frames you want to skip
    frame_skip = (fps//rfps)-1
    # a variable to keep track of the frame to be saved
    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if i > frame_skip - 1:
            frame_count += 1
            cv2.imwrite(dest + str(frame_count*frame_skip)+'.jpg', frame)
            i = 0
            continue
        i += 1

    cap.release()
    cv2.destroyAllWindows()

# video_location = '/content/Assault006_x264.mp4'
# get_frames(video_location)
```
